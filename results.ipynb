```python
with open('/content/drive/MyDrive/Colab Notebooks/Cost_contrastive_learning/CoST-main/training/mendeley/_/out.pkl','rb') as f:
  out = pkl.load(f)
pred = out[24]['raw'].flatten()[:1000]
gt = out[24]['raw_gt'].flatten()[:1000]

import matplotlib.pyplot as plt

plt.plot(gt, label = 'ground truth')
plt.plot(pred, label = 'prediction')
plt.xlabel('Index')
plt.ylabel('Voltage')
plt.legend(loc="upper right")
plt.show()
```
