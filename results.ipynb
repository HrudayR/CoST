#**Mendeley Dataset**


```python
import pickle as pkl
with open('/content/drive/MyDrive/Colab Notebooks/Cost_contrastive_learning/CoST-main/training/mendeley/_/eval_res.pkl','rb') as f:
  res = pkl.load(f)
print(res['ours'])
```

```python
with open('/content/drive/MyDrive/Colab Notebooks/Cost_contrastive_learning/CoST-main/training/mendeley/_/out.pkl','rb') as f:
  out = pkl.load(f)
pred = out[24]['raw'].flatten()[:1000]
gt = out[24]['raw_gt'].flatten()[:1000]

import matplotlib.pyplot as plt

plt.plot(gt)
plt.plot(pred)

plt.show()
```

```python
scinet_men_mse = [0.001	,0.0009	,0.0015	,0.0019	,0.0023]
scinet_men_mae = [0.0175 ,0.0186 ,0.0245 ,0.0294 ,0.0338]
cost_men_mse =[]
cost_men_mae = []
for val in list(res['ours'].keys()):
  cost_men_mse.append(res['ours'][val]['raw']['MSE'])
  cost_men_mae.append(res['ours'][val]['raw']['MAE'])
```

```python
# MSE Plot
plt.plot([24,48,96,168,336], scinet_men_mse, label='scinet')
plt.plot([24,48,96,168,336], cost_men_mse, label = 'COsT')
plt.legend(loc="upper left")
plt.show()
```

```python
# MAE Plot
plt.plot([24,48,96,168,336], scinet_men_mae, label='scinet')
plt.plot([24,48,96,168,336], cost_men_mae, label = 'COsT')
plt.legend(loc="upper left")
plt.show()
```
